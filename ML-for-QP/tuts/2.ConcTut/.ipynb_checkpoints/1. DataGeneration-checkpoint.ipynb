## Generating the Dataset

```python
import numpy as np
import pandas as pd
from numpy import linalg as LA
import matplotlib.pyplot as plt
from scipy import linalg as sLA
import numpy.random as r
import pandas as pd

np.random.seed(2022)
```

## Naive Approach to Generating Random Unitaries

A unitary matrix $U$ is a matrix that satisfies:
$$ UU^{\dagger} = U^{\dagger}U = I$$
and we know that it can be generated by a Hermitian matrix as follows:
$$ U = \mathrm{exp}(iM),\ U \in \mathrm{U}(N)\ \mathrm{and}\ M = M^{\dagger} \in \mathcal{M}_{N}(\mathbb{C})$$

So, if we can generate random Hermitian matrices the problem might be solved. But, how do we generate random Hermitian matrices?

Take any two random real $N \times N$  matrices $A,B$ then the matrix

$$H = (A+A^T)+i(B-B^T)$$

is a complex and random Hermitian matrix.

We can prove this as follows:
$$
\begin{align}
H^\dagger & = (A+A^T)^\dagger - i (B-B^T)^{\dagger}\\
& = (A^T+A)-i(B^T-B)\\
& = (A+A^T)+(B-B^T)\\
& = H
\end{align}
$$

Let's try this approach!

```python
def randU_v1(n):
    a = r.rand(n,n)
    b = r.rand(n,n)
    return sLA.expm(1j*((a + a.T)+1j*(b-b.T))/np.sqrt(2))
```

```python
u = randU_v1(4)

# to check that this is a Unitary matrix
np.round(np.conjugate(u.T).dot(u),10)
```

But, are these unitary matrices really random? Are they uniformly sampled from $\mathrm{U}(N)$?
We will now show that they are not random. 

We use the fact that all the eigenvalues of unitary matrices are of the form $e^{i\theta_1},...,e^{i\theta_N}$. 

```python
num_matrices = 1000

result = np.zeros((num_matrices,4)) 

for i in range(num_matrices):
    # find the eignevalues of a randombly generated unitary
    ev = LA.eigvals(randU_v1(4))

    # find the theta of that eigenvalue 
    result[i] = np.arctan2(ev.imag, ev.real)

# flatten the array, it will now be just a list of all the eigenvalues
result = result.reshape((4*num_matrices))
```

```python
# plot the distribution of the arrays
n_bins = 20
plt.hist(result, n_bins)
plt.show()
```

As we can see, the eigenvalues are not uniformly distributed. The probability of obtaining each eigenvalue is no where near equally likely. This means that our unitaries are not uniformly distributed.

## Using the Haar Measure to Generate Random Unitaries

```python
# Code taken from How to Generate Random Matrices from the Classical Compact Groups
# by Francesco Mezzadri

# returns a random matrix distributed with Haar measure.
def randU_haar_measure(n):
  # Generate an NxN with complex numbers a+bi where a and b are normally distributed with mean 0 and variance 1 (this is sampling from the distribution known as the Ginibre ensemble).
    z = (r.randn(n,n) + 1j*r.randn(n,n))/ np.sqrt(2.0)

    # Find Q R decomposition of matrix where Q is a unitary matrix and R is an upper triangular, invertible  matrix
    q,s = LA.qr(z)

    # Find the diagonal matrix if R 
    d = np.diagonal(s)

    # Divide each element of the diagonal by its absolute value 
    ph = d/np.absolute(d)
 
    # The Haar Random matrix is q x ph
    q = np.multiply(q,ph,q)
    return q
```

```python
num_matrices = 1000

result = np.zeros((num_matrices,4)) 

for i in range(num_matrices):
    # find the eignevalues of a randombly generated unitary
    ev = LA.eigvals(randU_haar_measure(4))

    # find the theta of that eigenvalue 
    result[i] = np.arctan2(ev.imag, ev.real)

# flatten the array, it will now be just a list of all the eigenvalues
result = result.reshape((4*num_matrices))
```

```python
# plot the distribution of the eigenvalues
n_bins = 20
plt.hist(result, n_bins)
plt.show()
```

## Generating the Dataset for our Regression Problem 

```python
def concurrence(x):
    return 2*np.abs(x[0]*x[3]-x[1]*x[2])
```

We will store the unitary as its action on a 2 qubit state in a uniform superposition. 

$$ 
\begin{align}
|\psi\rangle & = U (\frac{1}{2}|00\rangle + |01\rangle + |10\rangle + |11\rangle) 
& = 
\begin{pmatrix}
a \\
b \\
c \\
d 
\end{pmatrix}
\end{align}
$$

with $a,...,d \in \mathbb{C}$ and $a = a_1 + i a_2$ etc.

Basically, we have 8 columns storing the 8 values needed to represent the state and then we have 1 column storing the concurrence of the state. 

```python
num_points = 2000

# create an array with N rows and 9 columns 
data = np.zeros((num_points,9))

for row in range(num_points):
    # generate a random unitary using the Haar measure and then act that unitary on a 2-qubit state 
    # in uniform superposition
    v= np.dot(randU_haar_measure(4),np.array([1,1,1,1])/2)
    for col in range(0,8,2):
        data[row,col] = v[0].real
        data[row,col+1] = v[0].imag
    data[row,8] = concurrence(v)
```

```python
data
```

Now, we can store the data in a DataFrame.

```python
df = pd.DataFrame(data, columns=['a1', 'a2','b1','b2','c1', 'c2','d1', 'd2','conc'])
df.to_csv('data/states_reg.csv', index=False)
```

We can also look at the distribution of the concurrence values. 

```python
n_bins = 20
plt.hist(df['conc'].values, n_bins)
plt.show()
```
